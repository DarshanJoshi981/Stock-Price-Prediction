```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_datareader as pdr
```

# EDA on MRF dataset

```python
mrf = pdr.DataReader('MRF.NS',data_source='yahoo',start='2010-01-01',end='2020-01-01')
mrf.head(10)
```

```python
mrf.shape
```

```python
mrf.info()
```

```python
mrf.describe().T
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(mrf.corr(),cmap='YlGnBu',annot=True)
```

```python
#USED FOR INTRA-


mrf1 = mrf.copy()
mrf1['open-high'] = mrf1['Open']-mrf1['High']
mrf1['open-low'] = mrf1['Open'] - mrf1['Low']
mrf1['close-high'] = mrf1['Close']-mrf1['High']
mrf1['close-low'] = mrf1['Close'] - mrf1['Low']
mrf1['high-low'] = mrf1['High'] - mrf1['Low']
mrf1['open-close'] = mrf1['Open'] - mrf1['Close']
mrf1.head()
```

```python
mrf2 = mrf1.copy()
mrf2 = mrf2.drop(['Open','High','Low','Close', 'Adj Close'],axis=1)
plt.figure(figsize=(7,5))
sns.heatmap(mrf2.corr(),cmap='RdBu',annot=True)
```

```python
cols = ['High','Open','Low','Close','Adj Close','Volume']
i=0
while i<6:
    fig = plt.figure(figsize=[20,5])
    sns.set_theme(style='darkgrid')
    plt.subplot(1,2,1)
    sns.lineplot(data=mrf,x='Date',y=cols[i])
    plt.xlabel('Year')
    i += 1
    
    plt.subplot(2,2,2)
    sns.lineplot(data=mrf,x='Date',y=cols[i])
    plt.xlabel('Year')
    i += 1
    
    plt.subplot(3,2,1)
    sns.lineplot(data=mrf,x='Date',y=cols[i])
    i += 1
    
    plt.show()
```

```python
title = 'The MRF stock price'
ylabel='Price'
ax = mrf[['Open','Close']].plot(figsize=(12,6),title=title)
ax.autoscale(axis='x',tight=True)
ax.set(ylabel=ylabel);
```

```python
mrf3 = mrf.copy()
mrf3.drop('Volume',axis=1,inplace=True)
plt.figure(figsize=(20,10))
sns.lineplot(data=mrf3.loc['2015-01-01':'2020-12-31'])
plt.title('MRF Stocks in 2010 - 2015')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(data=mrf['Volume'].loc['2019-01-01':])
plt.title('MRF stock volumne for year 2019')
plt.xlabel('Year-Months')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
mrf3.loc['2010-12-12':'2019-12-31'].plot(kind='area', ax=ax, stacked=False)
plt.title('MRF Stocks in 2010 - 2019')
plt.xlabel('Year')
plt.show()
```

```python
title = 'Yearly Mean Closing Price'
mrf['Open'].resample('A').mean().plot.bar(title=title,color=['#7CA982'] , figsize=(15,5));
```

```python
import plotly.graph_objects as go
from datetime import datetime


fig = go.Figure(data=[go.Candlestick(x=mrf.index,
                open=mrf['Open'],
                high=mrf['High'],
                low=mrf['Low'],
                close=mrf['Close'])])

fig.show()
```

### We cannot observe clear patterns by simply comparing the Open price and the Close price. This is why we explore the dataset by adding new variables :
#### This variable measures the gap between the Close price and the day-after Open price 

```python
mrf4 = mrf.copy()
mrf4 = mrf.loc['2018-12-10' : '2018-12-17']
mrf['delta_open_close_day_before_%']=((mrf['Open']-mrf['Close'].shift(1))/mrf['Close'].shift(1))*100
mrf4
```

```python
order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
sns.boxplot(x=mrf.index.day_name(), y=mrf['delta_open_close_day_before_%'], order=order);
```
