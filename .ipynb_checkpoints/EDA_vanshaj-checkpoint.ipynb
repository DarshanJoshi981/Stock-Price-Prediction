```python
import numpy as np
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
```

```python
def avgAnnualPrice(data):
    cmpName = data["Company Name"].unique()[0]
    df = data.reset_index()
    df.drop(["Adj Close"], axis=1)
    # yearList = data["Date"].dt.year.unique()
    resultDf = df.groupby(df["Date"].dt.year)["Close"].mean()
    f = plt.figure()
    f.set_figwidth(10)
    f.set_figheight(7)
    plt.bar(resultDf.keys(), resultDf.values, width = 0.5, color = "Grey")
    plt.xlabel("Years", fontweight ='bold', fontsize = 8)
    plt.ylabel("Annual Average Closing Price", fontweight ='bold', fontsize = 8)
    plt.title("Annual Avg Price of " + cmpName, fontweight ='bold', fontsize = 10)
    plt.show()


```

```python
df1 = pdr.DataReader('MRF.NS',data_source='yahoo',start='2010-01-01',end='2020-01-01')
df1['Company Name'] = 'MRF'
avgAnnualPrice(df1)
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center',
            Bbox = dict(facecolor = 'red', alpha =.8))

def annualPercentageIncrease(data):
    perIncrease = []
    cmpName = data["Company Name"].unique()[0]
    df = data.reset_index()
    df.drop(["Adj Close"], axis=1)
    yearList = ["2010-11", "2011-12", "2012-13", "2013-14", "2014-15", "2015-16", "2016-17", "2017-18", "2018-19", "2019-20"]
    resultDf = df.groupby(df["Date"].dt.year)["Close"].mean()
    for i in range(10):
        per = round((resultDf.iloc[i+1] - resultDf.iloc[i]) / resultDf.iloc[i] * 100,2)
        perIncrease.append(per)
    


    f = plt.figure()
    f.set_figwidth(10)
    f.set_figheight(7)
    plt.bar(yearList, perIncrease, width = 0.5, color = "Gray")
    addlabels(yearList, perIncrease)
    plt.xlabel("Years", fontweight ='bold', fontsize = 8)
    plt.ylabel("Percentage Increase", fontweight ='bold', fontsize = 8)
    plt.title("Percentage increase in succesive years of " + cmpName, fontweight ='bold', fontsize = 10)
      
    plt.show()
```

```python
annualPercentageIncrease(df1)
```
