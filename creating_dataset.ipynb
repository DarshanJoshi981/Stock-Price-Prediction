```python
import numpy as np
import pandas as pd
import pandas_datareader as pdr
```

```python
df1 = pdr.DataReader('MRF.NS',data_source='yahoo',start='2010-01-01',end='2020-01-01')
df1['Company Name'] = 'MRF'
df1
```

```python
df2 = pdr.DataReader('M&M.NS',data_source='yahoo',start='2010-01-01',end='2020-01-01')
df2['Company Name'] = 'M&M'
df2
```

```python
df3 = pdr.DataReader('TATAMOTORS.BO',data_source='yahoo',start='2010-01-01',end='2020-01-01')
df3['Company Name'] = 'TATAMOTORS'
df3
```

```python
df4 = pdr.DataReader('HDFCBANK.NS',data_source='yahoo',start='2010-01-01',end='2020-01-01')
df4['Company Name'] = 'HDFCBANK'
df4
```

```python
df5 = pdr.DataReader('RELIANCE.NS',data_source='yahoo',start='2010-01-01',end='2020-01-01')
df5['Company Name'] = 'RELIANCE'
df5
```

```python
data = pd.concat([df1,df2,df3,df4,df5], ignore_index=True)
```

```python
data
```

```python
data.to_csv('stock_data.csv')
```
